```python
# Obecně je BMI (Body Mass Index) míra, která vyčísluje obezitu u lidí.

# Rovnice výpočtu BMI: váha / výška2


# Váha se udává v kilogramech a výška v metrech. Hodnoty, které z tohoto vzorečku dostaneme, škálujeme podle této tabulky:
# Hodnota	Význam
# < 18,5	Podvýživa
# 18,5 – 25	Zdravá váha
# 25 – 30	Mírná nadváha
# 30 – 40	Obezita
# 40 <	Těžká obezita

# V této úloze vytvoř program, který získá BMI uživatele Martin, který měří 200 centimerů a váží 80 kilogramů:

#     Vytvoř proměnné jmeno, vaha, vyska a zadej do nich hodnoty,
jmeno = ("Martin")
vaha = int(120)
vyska = int(2)

#     vytvoř proměnnou bmi a přiřaď k ní vzorec, pomocí proměnných vaha, vyska a aritmetického operátoru na druhou ,
bmi = vaha / (vyska ** 2)
#     vytvoř proměnnou kategorie, do které uložíš název kategorie odpovídající hodnotě BMI,
kategorie = (bmi)
#     vypiš výsledek do věty, jak je uvedeno níže.
if bmi > 40:
    kategorie = "těžká obezita"
elif bmi > 30:
    kategorie = "obezita"
elif bmi > 25:
    kategorie = "mírná nadváha"
elif bmi > 18.5:
    kategorie = "zdravá váha"
else :
    kategorie = "podvýživa"
# Martin tvé BMI je 20.0, což spadá do kategorie zdravá váha.
print(f"{jmeno} tvé BMI je {bmi} což spadá do kategorie {kategorie}")
```

```python
# V této úloze vytvoř program, který bude pracovat s palindromy.

# Palindromy jsou taková slova, která se čtou stejně zepředu i pozpátku. Ty napíšeš takový program, který je rozpozná.

# Při řešení nepoužívej funkci input.

# Postup bude vypadat následovně:

#     Vytvoř list zadana_slova, který obsahuje hodnoty: ["jaro", "jej", "kolo", "aha"]
zadana_slova = ["jaro", "jej", "kolo", "aha"]
#     vyber libovolné slovo z listu zadana_slova a ulož jej do proměnné slovo (pomocí indexu),
slovo = (zadana_slova[0])
#     pokud se jedná o palindrom, vypsat "Slovo '<slovo>' je palindrom!",
if slovo == (slovo[::-1]):
    print(f"Slovo '{slovo}' je palindrom!")
#     pokud se nejedná o palindrom, vypsat - "Slovo '<slovo>' není palindrom".
else :
    print(f"Slovo '{slovo}' není palindrom")



# Ukázka průběhu

# Zadej slovo: jaro
# Slovo 'jaro' není palindrom.

# Zadej slovo:jej
# # Slovo 'jej' je palindrom!.
```

```python
# Zadání

# V této úloze vytvoř program, který:

#     Vytvoř list zamestnanci, který bude obsahovat stringy 'František', 'Anna', 'Jakub', 'Klára',
zamestnanci = ['František', 'Anna', 'Jakub', 'Klára']
#     vypiš obsah zamestnanci za větu 'Zaměstnanci na začátku: ',
print(f"Zaměstnanci na začátku: {zamestnanci}")
#     vytvoř kopii listu zamestnanci a pojmenuj proměnnou zamestnanci_a,
zamestnanci_a = zamestnanci.copy()
#     přidej do listu zamestnanci_a jména 'Bruno' a 'Anežka',
zamestnanci_a.append('Bruno')
zamestnanci_a.append('Anežka')
#     vypiš obsah listu zamestnanci_a za string 'Nová jména přidána: ',
print(f"Nová jména přidána: {zamestnanci_a}")
#     vytvoř kopii listu zamestnanci a pojmenuj proměnnou zamestnanci_b,
zamestnanci_b = zamestnanci.copy()
#     vlož jméno 'Bruno' na index 1,
zamestnanci_b.insert(1, 'Bruno')
#     vypiš obsah proměnné zamestnanci_b za string: 'Nová jména vložena:'.
print(f"Nová jména vložena: {zamestnanci_b}")
# Řešení

# Zaměstnanci na začátku: ['František', 'Anna', 'Jakub', 'Klára']
# Nová jména přidána: ['František', 'Anna', 'Jakub', 'Klára', 'Bruno', 'Anežka']
# Nová jména vložena: ['František', 'Bruno', 'Anna', 'Jakub', 'Klára']
```

```python
# V této úloze vytvoř program, který:

#     Na záčatku nachystá tyto proměnné:
#     Ukázka kódu

# ZKOPÍROVAT KÓD

kandidati = ['Bruno', 'Anežka']

zamestnanci = ['František', 'Bruno', 'Anna', 'Jakub', 'Klára']

#     vytvoří proměnnou bez_bruna, kam přiřadí hodnoty z proměnné kandidati,
bez_bruna = kandidati.copy()
#     odstraní z listu bez_bruna string 'Bruno',
bez_bruna.remove('Bruno')
#     vypíše obsah proměnné bez_bruna za string: 'Bruno odstraněn:',
print(f'Bruno odstraněn:{bez_bruna}')
#     zopakuje a uloží třikrát obsah listu bez_bruna do proměnné opakovani_kandidati,
opakovani_kandidati = bez_bruna * 3
#     vypíše obsah proměnné opakovani_kandidati za string: 'Opakování kandidáti:',
print(f'Opakování kandidáti: {opakovani_kandidati}')
#     spojí list zamestnanci s listem opakovani_kandidati do nového listu spojeni_zamestnanci,
spojeni_zamestnanci = zamestnanci + opakovani_kandidati
#     vypíše obsah nového listu spojeni_zamestnanci za string: 'Spojeni zaměstnanci:'.
print(f'Spojení zaměstnanci: {spojeni_zamestnanci}')

# Řešení

# Bruno odstraněn: ['Anežka']
# Opakování kandidáti: ['Anežka', 'Anežka', 'Anežka']
# Spojení zaměstnanci: ['František', 'Bruno', 'Anna', 'Jakub', 'Klára', 'Anežka',
# 'Anežka', 'Anežka']
```

```python
# Tvým úkolem bude ověřit správnost prvního písmene každého dne v týdnu.

# Postup bude vypadat následovně:

#     Zapiš tyto vstupní proměnné:
#     Ukázka kódu

# ZKOPÍROVAT KÓD

vstupni_cisla = [1, 2, 3, 4, 5, 6, 7]

vstupni_pismena = ["p", "ú", "s", "č", "p", "s", "n"]

tyden = ('pondělí', 'úterý', 'středa', 'čtvrtek', 'pátek', 'sobota', 'neděle')

#     vytvoř proměnnou cislo_dne a zapiš do ní hodnotu 3,
cislo_dne = 3
#     ověř, jestli hodnota v cislo_dne je v listu vstupni_cisla,
if cislo_dne in vstupni_cisla:
    print("Spravná vstupní hodnota!")

#     pokud není, vypiš výstup podle vzoru níže,
#     pokud je, vypiš zprávu "Správná vstupní hodnota!"
    den_tydne = tyden[cislo_dne - 1]
    if den_tydne.startswith(vstupni_pismena[cislo_dne - 1]):
        print("Správné písmeno")

    else:
        print("Špatné písmeno!")
#     .. dále vytvoř proměnnou den_tydne a pomocí upravené hodnoty v proměnné cislo_dne, 
#               naindexuj z proměnné tyden požadovaný den (př. 1 --> "pondělí", 2 --> "úterý"),
#      ověř, jestli vybraný str v proměnné den_tydne má stejné počáteční písmeno, 
#               jako jsou v proměnné vstupni_pismena (ověř opět pomocí upraveného indexu),

#     .. pokud se znaky shodují vypiš "Správné písmeno",
#     .. pokud se znaky neshodují vypiš "Špatné písmeno".

else:
    print("Špatná vstupní hodnota!")
# Ukázka průběhu


# # cislo_dne = 3 --> "středa" == "s"
# Správná vstupní hodnota!
# Správné písmeno

# # cislo_dne = 9 
# Špatná vstupní hodnota!
```
